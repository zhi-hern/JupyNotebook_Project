## **_Evaluate Sales Performance_** 
### Calculate Gross Profit Margin (GPM) which indicates the efficiency of production and pricing.



- **Assumption 1: Total Revenue = `'Sales'` in data**
- **Assumption 2: Gross Profit = `'Profit'` in data**
- Gross Profit Margin (GPM) = Profit/Sales x 100%
- Top 10 companies with the highest GPM were identified.
- Top 10 industries with the highest GPM were identified.
- Top companies in these industries were determined.
<br>

Ref: https://ramp.com/blog/how-to-measure-company-financial-performance 


```python
GPM = df[['Name','Sales (in Billions)','Profit (in Billions)','Industry','Country']]
GPM.insert(3,'Gross Profit Margin',df['Profit (in Billions)']/df['Sales (in Billions)']*100,True)
GPM.head()
```

```python
GPM = GPM.sort_values(by='Gross Profit Margin',ascending=False).reset_index()
GPM.drop('index',axis=1,inplace=True)
GPM.insert(0,'Rank (in GPM)', GPM.index+1, False)
GPM.head()
```

## **_Evaluate Asset Management_**
### Calculate Asset Turnover Ratio (ATR) which measures the efficiency with which a company uses its assets to produce sales.

- **Assumption 1: Net Sales = `'Sales'` in data**
- **Assumption 2: Average Total Assets = `'Assets'` in data**
- Asset Turnover Ratio (ATR) = Sales/Assets

Ref: https://corporatefinanceinstitute.com/resources/accounting/asset-turnover-ratio/

```python
ATR = df[['Name','Sales (in Billions)','Assets (in Billions)','Industry','Country']]
ATR.insert(3,'Asset Turnover Ratio',df['Sales (in Billions)']/df['Assets (in Billions)'],True)
ATR.head()
```

### Top 10 companies that have the highest Asset Turnover Ratio

```python
ATR = ATR.sort_values(by='Asset Turnover Ratio',ascending=False).reset_index()
ATR.drop('index',axis=1,inplace=True)
ATR.insert(0,'Rank (in ATR)', GPM.index+1, False)
ATR.head(10)
```

### Top 10 industries that have the highest Asset Turnover Ratio

```python
ATR.groupby(by='Industry').agg({'Industry':'count','Asset Turnover Ratio':'mean'}).sort_values(by='Asset Turnover Ratio',ascending=False)\
                         .rename(columns={'Industry':'Count','Asset Turnover Ratio':'Avg Asset Turnover Ratio'}).head(10)
```

- Findings: Industries that have high Asset Turnover Ratio usually involve products that are well-accessible by the public.

## **_Profitability Analysis_**
### Calculate Return on Assets (ROA) which indicates how effectively a company utilizes its assets to generate profit, reflecting operational efficiency and asset management.


- **Assumption 1: Net Income = `'Profit'` in data**
- **Assumption 2: Total Assets = `'Assets'` in data**
- Return on Assets (ROA) = Profit/Assets

Ref: https://www.investopedia.com/terms/r/returnonassets.asp

```python
ROA = df[['Name','Profit (in Billions)','Assets (in Billions)','Industry','Country']]
ROA.insert(3,'Return on Assets',df['Profit (in Billions)']/df['Assets (in Billions)'],True)
ROA.head()
```

### Top 10 companies that have the highest Return on Assets 

```python
ROA = ROA.sort_values(by='Return on Assets',ascending=False).reset_index()
ROA.drop('index',axis=1,inplace=True)
ROA.insert(0,'Rank (in ROA)', GPM.index+1, False)
ROA.head(10)
```

### Top 10 industries that have the highest Return on Assets 

```python
ROA.groupby(by='Industry').agg({'Industry':'count','Return on Assets':'mean'}).sort_values(by='Return on Assets',ascending=False)\
                         .rename(columns={'Industry':'Count','Return on Assets':'Avg Return on Assets'}).head(10)
```
